```python
import os

def rename_files_in_folder(folder_path):
    # List all files in the folder
    files = os.listdir(folder_path)

    for filename in files:
        # Construct the old file path
        old_file_path = os.path.join(folder_path, filename)

        # Check if it is a file (and not a directory)
        if os.path.isfile(old_file_path):
            # Construct the new file name
            new_filename = "preprocessed_" + filename
            # Construct the new file path
            new_file_path = os.path.join(folder_path, new_filename)

            # Rename the file
            os.rename(old_file_path, new_file_path)
            print(f'Renamed: {old_file_path} to {new_file_path}')

# Specify the folder path
folder_path = '/content/drive/MyDrive/OriginalData/Epileptic'
rename_files_in_folder(folder_path)
```

```python
import os

def rename_files_in_folder(folder_path):
    # List all files in the folder
    files = os.listdir(folder_path)

    for filename in files:
        # Construct the old file path
        old_file_path = os.path.join(folder_path, filename)

        # Check if it is a file (and not a directory)
        if os.path.isfile(old_file_path):
            # Check if the file has the extension .nii.gz
            if filename.endswith('.nii.gz'):
                # Split the file name and its special extension
                name = filename[:-7]  # Remove the last 7 characters (".nii.gz")
                # Construct the new file name
                new_filename = name + "_roi.nii.gz"
                # Construct the new file path
                new_file_path = os.path.join(folder_path, new_filename)

                # Rename the file
                os.rename(old_file_path, new_file_path)
                print(f'Renamed: {old_file_path} to {new_file_path}')

# Specify the folder path
folder_path = '/content/drive/MyDrive/OriginalData/Epileptic'
rename_files_in_folder(folder_path)
```

```python
import os

def undo_rename_files_in_folder(folder_path):
    # List all files in the folder
    files = os.listdir(folder_path)

    for filename in files:
        # Construct the old file path
        old_file_path = os.path.join(folder_path, filename)

        # Check if it is a file (and not a directory)
        if os.path.isfile(old_file_path):
            # Split the file name and its extension
            name, ext = os.path.splitext(filename)
            # Check if the file name ends with "_roi"
            if name.endswith('_roi'):
                # Construct the new file name by removing "_roi"
                new_name = name[:-4]  # Remove the last 4 characters ("_roi")
                new_filename = new_name + ext
                # Construct the new file path
                new_file_path = os.path.join(folder_path, new_filename)

                # Rename the file
                os.rename(old_file_path, new_file_path)
                print(f'Renamed: {old_file_path} to {new_file_path}')

# Specify the folder path
folder_path = '/content/drive/MyDrive/OriginalData/Epileptic'
undo_rename_files_in_folder(folder_path)
```
